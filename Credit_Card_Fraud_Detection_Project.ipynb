```python
  # percentage of total not fraud transaction
nfcount = 0
notfraud=df['Class']
for i in range(len(notfraud)):
   if notfraud[i]==0:
    nfcount = nfcount + 1

nfcount   
per_nf = (nfcount/len(notfraud))*100
print("percentage of total not fraud transaction in the transactionin the dataset:",per_nf)
```

```python
# percentage of total fraud transaction
fcount = 0
fraud=df['Class']
for i in range(len(fraud)):
   if fraud[i]==1:
    fcount = fcount + 1

fcount   
per_f = (fcount/len(fraud))*100
print("percentage of total fraud transaction in the transactionin the dataset:",per_f)
```

```python
  x= df.drop(['Class'], axis = 1)
  y= df['Class'] 
  xtrain,xtest, ytrain, ytest = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
lg = LogisticRegression()
lg.fit(xtrain,ytrain)
```

```python
y_pred = lg.predict(xtest)
accuracy = lg.score(xtest,ytest)
```

```python
cm = metrics.confusion_matrix(ytest,y_pred)
print(cm)
```

```python
print('Accuracy score of the Logistic regression model:' , accuracy*100,'%')
```
